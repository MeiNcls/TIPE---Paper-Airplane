```python
def remplace(s):
    return s.replace("," , ".").encode()

def lire_fichier(nom):
    """
    Cette fonction lit un fichier de données au format txt avec trois colonnes t,x
    Elle renvoie les trois colonnes t,x,y
    nom est une chaîne de caractères donnant le nom du fichier sans l'extension txt (string)
    """
    fichier=open(nom+".txt") # on ouvre un flux de fichier en lecture
    t,x,y=loadtxt( (remplace(s) for s in fichier),delimiter=';',skiprows=2,unpack=True)
    fichier.close()
    return t,x,y

def graph_trajectoire_brut(nom,nom_figure):
    """
    Prend un fichier texte (avec trois colonnes) en entrée et le nom de la figure qu'on veut tracer
    nom, nom_figure sont des chaines de caractères (nom donne le nom du fichier sans l'extension txt)
    Trace le graphe de la trajectoire de l'avion (y = f(x))
    """
    t,x,y = lire_fichier(nom)
    close(nom_figure)
    figure(nom_figure)
    plot(x,y)
    grid()
    xlabel('position selon ex')
    ylabel('position selon ey')
    show()
    return None
    
```

```python
graph_trajectoire('10_e4_donnees_pointage_calibre','essai 4, séance 10')
graph_trajectoire('10_e6_donnees_pointage_calibre','essai 6, séance 10')
graph_trajectoire('9_oiseau_donnees_pointage_calibre','essai oiseau, séance 9')
```

```python
#lissage des courbes
def lissage(tab,m):
    """
    Prend en entrée un tableau numpy tab et un nombre m 
    Renvoie tab_liss un tableau numpy de même taille que tab (pour l'instant en 1D)
    Renvoie le tableau lissé ie pour les valeurs où c'est possible la moyenne avec les points autour
    """
    l = len(tab)
    n=m//2
    tab_liss = zeros(l)
    for i in range (n):
        tab_liss[i]=tab[i]
        tab_liss[-i-1] =tab[-i-1]
    for i in range (n,l-n):
        t = tab[i-n:i+n] # attention c pas vrmt symétrique peut être l'erreur
        s = sum(t)
        tab_liss[i] = s/(n*2)
    return tab_liss

def graph_traj_liss(nom,nom_figure,m):
    """
    Prend en entrée le nom d'un fichier (dont ont extrait 3 colonnes t,x,y) sans l'extension txt (str), 
    ainis que le nom donné à la figure sans lissage (str)
    et m le nombre de points sur lequel on veut lisser (faire la moyenne). (int a priori même si ça peut marcher avec float)
    De fait on lissera sur (m//2)*2.
    Renvoie la trajectoire lissée (x et y ont été lissés par la fonction de lissage)
    """
    t,x,y = lire_fichier(nom)
    x_liss = lissage(x,m)
    y_liss = lissage(y,m)
    close(nom_figure+' lissé')
    figure(nom_figure+' lissé')
    plot(x_liss,y_liss)
    grid()
    xlabel('position selon ex')
    ylabel('position selon ey')
    show()
```

```python
graph_traj_liss('10_e6_donnees_pointage_calibre','essai 6, séance 10',8)
graph_traj_liss('10_e4_donnees_pointage_calibre','essai 4, séance 10',6)
graph_traj_liss('9_oiseau_donnees_pointage_calibre','essai 6, séance 10',4)
```
